```python
from itertools import product
from fractions import Fraction
from collections import defaultdict
import matplotlib.pyplot as plt
```

```python
## SUMA
def P(A, Omega):
  P= Fraction(len(A),len(Omega))
  return P

## 1° Crear espacio muestral Omega
L= [i for i in range(1,7)] #Lista por comprensión

Omega= set(product(L, repeat=2))

S= {(i,j): i+j for i,j in Omega} #suma

# Crear diccionario por default de manera sintetica y arreglada
from collections import defaultdict
dS= defaultdict(set) #lista

for i,j in S.items():
  dS[j].add(i) #añadir al diccionario las llaves

leyS= {k:P(A,Omega) for k,A in dS.items()}

sum_probS = sum(leyS.values())
verificacionS = sum_probS == 1

print("Omega =", Omega, "\n", 'Cardinalidad =', len(Omega))
print(leyS)
print("Suma de las probabilidades =", sum_probS)
print("¿La suma es igual a 1? =", verificacionS)
```

Graficando

```python
suma = list(leyS.keys())
probabilidades = [float(leyS[k]) for k in suma]  # Convertir a flotante para la gráfica

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.bar(suma, probabilidades, color='blue', edgecolor='black')

# Añadir títulos y etiquetas
plt.title('Distribución de Probabilidades de la Suma de Dos Dados')
plt.xlabel('Suma de las Caras')
plt.ylabel('Probabilidad')
plt.xticks(suma)  # Mostrar cada valor de suma en el eje x

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

* La diferencia de las caras es mayor que 1

```python
# DIFERENCIA

def P(Ad, Omegad):
    return Fraction(len(Ad), len(Omegad))

# 1° Crear espacio muestral Omega
Ld = [i for i in range(1, 7)]
Omegad = set(product(Ld, repeat=2))

# Crear diccionario con la diferencia de las caras de los dados
D = {(i, j): abs(i - j) for i, j in Omegad}

# Crear diccionario por default para almacenar las combinaciones según la diferencia
dD = defaultdict(set)

for (i, j), diff in D.items():
    dD[diff].add((i, j))

# Calcular la ley de probabilidad para cada diferencia posible
leyD = {k: P(Ad, Omegad) for k, Ad in dD.items()}

# Verificar que la suma de las probabilidades sea igual a 1
suma_probad = sum(leyD.values())
verificaciondd = suma_probad == 1

print("Omega =", Omegad, "\n", 'Cardinalidad =', len(Omegad))
print("Suma de las probabilidades =", suma_probad)
print("¿La suma es igual a 1? =", verificaciondd)
```

```python
sumad = list(leyD.keys())
p_d = [float(leyD[k]) for k in sumad]  # Convertir a flotante para la gráfica

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.bar(sumad, p_d, color='green', edgecolor='black')

# Añadir títulos y etiquetas
plt.title('Distribución de Probabilidades de diferencia de Dos Dados')
plt.xlabel('Diferencia de las Caras')
plt.ylabel('Probabilidad')
plt.xticks(sumad)  # Mostrar cada valor de suma en el eje x

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

Ahora filtramos las diferencias que sea mayores a 1

```python
dD_filtro = {k: Ad for k, Ad in dD.items() if k > 1}

# Calcular la ley de probabilidad para las diferencias filtradas
leyD_filtro = {k: P(Ad, Omegad) for k, Ad in dD_filtro.items()}

# Verificar que la suma de las probabilidades filtradas sea igual a 1
sum_p_filtro = sum(leyD_filtro.values())
ver_filtro = sum_p_filtro == 1

print("Omega =", leyD_filtro, "\n", 'Cardinalidad =', len(leyD_filtro))
print("Suma de las probabilidades filtradas =", sum_p_filtro)
print("¿La suma es igual a 1 para diferencias mayores que 1? =", ver_filtro)
```

```python
sumaf = list(leyD_filtro.keys())
p_f = [float(leyD_filtro[k]) for k in sumaf]  # Convertir a flotante para la gráfica

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.bar(sumaf, p_f, color='green', edgecolor='black')

# Añadir títulos y etiquetas
plt.title('Distribución de Probabilidades de diferencia >1')
plt.xlabel('Diferencia de las Caras')
plt.ylabel('Probabilidad')
plt.xticks(sumaf)  # Mostrar cada valor de suma en el eje x

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

* La suma de caras sea par

```python
def P(Ap, Omegap):
    return Fraction(len(Ap), len(Omegap))

# 1° Crear espacio muestral Omega
Lp = [i for i in range(1, 7)]
Omegap = set(product(Lp, repeat=2))
print("Omega suma par =", Omegap, "\n", 'Cardinalidad =', len(Omegap))

# Crear diccionario con la suma de las caras de los dados
Sp = {(i, j): i + j for i, j in Omegap}

# Crear diccionario por default para almacenar las combinaciones según la suma de las caras
dSp = defaultdict(set)

for (i, j), suma in Sp.items():
    dSp[suma].add((i, j))

# Filtrar combinaciones donde la suma de las caras es par
dS_par = {k: Ap for k, Ap in dSp.items() if k % 2 == 0}

# Calcular la ley de probabilidad para la suma par
leyS_par = {k: P(Ap, Omegap) for k, Ap in dS_par.items()}

# Verificar que la suma de las probabilidades de las sumas pares sea igual a 1
suma_prob_par = sum(leyS_par.values())
ver_par = suma_prob_par == 1

# Verificar la suma de las probabilidades en el espacio muestral completo
leySp = {k: P(Ap, Omegap) for k, Ap in dSp.items()}
sum_prob_c = sum(leySp.values())
ver_c = sum_prob_c == 1

print("\nLey de probabilidades para sumas pares:", leyS_par)
print("Suma de las probabilidades (pares) =", suma_prob_par)
print("¿La suma es igual a 1 para sumas pares? =", ver_par)

print("\nLey de probabilidades completa:", leySp)
print("Suma de las probabilidades (completa) =", sum_prob_c)
print("¿La suma es igual a 1 para el espacio muestral completo? =", ver_c)
```

```python
## SUMA COMPLETA
sumapc = list(leySp.keys())
p_pc = [float(leySp[k]) for k in sumapc]  # Convertir a flotante para la gráfica

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.bar(sumapc, p_pc, color='purple', edgecolor='black')

# Añadir títulos y etiquetas
plt.title('Distribución de Probabilidades de sumas pares')
plt.xlabel('Suma de las Caras')
plt.ylabel('Probabilidad')
plt.xticks(sumapc)  # Mostrar cada valor de suma en el eje x

# Mostrar el gráfico
plt.grid(True)
plt.show()
```
